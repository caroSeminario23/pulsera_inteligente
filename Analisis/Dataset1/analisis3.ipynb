```python
import pandas as pd

# Leer el archivo csv
data1 = pd.read_csv('Dataset_medio_limpio_filtrado.csv')
data1
```

```python
# visualizar los datos unicos de la columna disease
data1['Disease'].unique()
```

## A seleccionar:
- Influenza
- Asthma
- Depression
- Bronchitis
- Hypertensive Heart Disease
- Migraine

- Diabetes

```python
# Crear un nuevo dataset con los registros que contengan la enfermedad 'Hypertension' o 'Diabetes'
data2 = data1[(data1['Disease'] == 'Influenza') | (data1['Disease'] == 'Asthma')| (data1['Disease'] == 'Migraine') | (data1['Disease'] == 'Pneumonia') | (data1['Disease'] == 'Diabetes') | (data1['Disease'] == 'Eating Disorders') | (data1['Disease'] == 'Hypertension')]
```

```python
data2
```

```python
# Borrar los registros con id 1, 50-53
data3 = data2.drop([1,50,51,52,53])
data3
```

```python
# Exportar el dataset a un archivo csv
data3.to_csv('Dataset_medio_limpio_filtrado2.csv', index=False)
```
