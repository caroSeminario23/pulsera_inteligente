```python
# Generar un gráfico que muestre la relación entre las columnas 'Prognosis_code' y las demas
for i in data2.columns[:-2]:
    sintoma = data2[i]
    enfermedad = data2['Prognosis_code']

    plt.bar(enfermedad, sintoma)
    plt.xlabel('Prognosis')
    plt.ylabel(i)
    plt.title(i + ' vs Enfermedad')

    # Configurar las etiquetas del eje x para que aparezcan de uno en uno
    plt.xticks(range(min(enfermedad), max(enfermedad)+1))
    # Rotar las etiquetas del eje x
    plt.xticks(rotation=90)
    plt.show()
```

- Los síntomas que se encuentran más frecuentes en la mayoría de enfermedades contempladas son:
    - chills (escalosfríos)
    - vomiting (vómito)
    - fatigue (fatiga)
    - high fever (fiebre alta)
    - headache (dolor de cabeza)
    - nausea (náuseas)
    - abdominal pain (dolor abdominal)
- Por ello, es necesario un diagnóstico diferencial que permita contrastar la enfermedad probable en base a otros síntomas que la caractericen.

```python
for i in data2['Prognosis_code'].unique():
    data_prognosis = data2[data2['Prognosis_code'] == i]

    # Guardar los sintomas con valores igual a uno para la prognosis 1
    symptoms_prognosis = data_prognosis.iloc[:, :-2].sum()

    # Crear el gráfico de barras
    plt.figure(figsize=(10, 6))
    plt.bar(symptoms_prognosis.index, symptoms_prognosis.values)
    plt.xlabel('Síntomas')
    plt.ylabel('Frecuencia')
    plt.title('Síntomas asociados a la prognosis ' + str(i))
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Calcular el número de gráficos
n = len(data2['Prognosis_code'].unique())

# Crear una figura para todos los gráficos
plt.figure(figsize=(10, 6))

# Calcular el número de filas y columnas para los subplots
rows = int(n**0.5)
cols = n // rows
cols += n % rows

# Índice para los subplots
plot_number = 1

for i in data2['Prognosis_code'].unique():
    data_prognosis = data2[data2['Prognosis_code'] == i]

    # Guardar los sintomas con valores igual a uno para la prognosis 1
    symptoms_prognosis = data_prognosis.iloc[:, :-2].sum()

    # Crear el gráfico de barras en un subplot
    plt.subplot(rows, cols, plot_number)
    plt.bar(symptoms_prognosis.index, symptoms_prognosis.values)
    plt.xlabel('Síntomas')
    plt.ylabel('Frecuencia')
    plt.title('Síntomas asociados a la prognosis ' + str(i))
    plt.xticks(rotation=90)

    # Incrementar el índice para el próximo subplot
    plot_number += 1

# Mostrar todos los gráficos
plt.tight_layout()
plt.show()
```

Como se puede visualizar, ninguna enfermedad probable se identifica por los mismos síntomas
